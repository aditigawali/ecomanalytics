```python
import datetime as dt
ecom_final.date_of_purchase.max()
```

```python
recency_date=dt.datetime(2017,12,20)
r=ecom_final.groupby('CustomerID',as_index=False).agg({'date_of_purchase': lambda x:(recency_date -x.max()).days})
```

```python
f=ecom_final.groupby('CustomerID',as_index=False).agg({'invoiceno': lambda x:(x.nunique())})
```

```python
m=ecom_final.groupby('CustomerID',as_index=False).agg({'Price' : lambda x :(x.sum())})
```

```python
m.drop('CustomerID',axis=1,inplace=True)
f.drop('CustomerID',axis=1,inplace=True)
```

```python
rmf_scores=pd.concat([r,f,m],axis=1)
```

```python
plt.figure(figsize=(10,8))
plt.hist(rmf_scores.Price,bins=30)
```

```python
rmf_scores.columns =['CustomerID','Recency','Frequency','Monetary']
rmf_scores
```

```python
rmf_scores['r_quartile'] = pd.qcut(rmf_scores['Recency'], 3, ['3','2','1'])

rmf_scores['m_quartile'] = pd.qcut(rmf_scores['Monetary'], 3, ['1','2','3'])
```

```python
rmf_scores.insert(6,'f_quartile',rmf_scores.Frequency)
```

```python
rmf_scores.f_quartile[rmf_scores.f_quartile<=4] =1
rmf_scores.f_quartile[(rmf_scores.f_quartile>1) & (rmf_scores.f_quartile <=20)] =2
rmf_scores.f_quartile[(rmf_scores.f_quartile>2)] =3
```

```python
rmf_scores['rfm'] =rmf_scores.r_quartile.astype(str)+ rmf_scores.f_quartile.astype(str) + rmf_scores.m_quartile.astype(str)
```

```python
rmf_scores=rmf_scores.astype({'r_quartile':'int64'})
rmf_scores=rmf_scores.astype({'m_quartile':'int64'})
```

```python
rmf_scores['rfmscore'] =rmf_scores[['r_quartile','m_quartile','f_quartile']].sum(axis=1)
```

```python
Log_trf_data = rmf_scores[['Recency','Frequency','Monetary']].apply(np.log, axis=1).round(2)
Log_trf_data
```

```python
def low_values(num):
    if num <=0:
        return 1
    else:
        return (num)
```

```python
Log_trf_data['Recency'] = [low_values(x) for x in Log_trf_data.Recency]
Log_trf_data['Frequency'] = [low_values(x) for x in Log_trf_data.Frequency]
Log_trf_data['Monetary']=[low_values(x) for x in Log_trf_data.Monetary] 
```

```python
Log_trf_data.drop(columns=('CustomerID'),inplace=True)
```

```python

Log_trf_data['CustomerID']=list(range(0,4324))
```

```python
Log_trf_data=Log_trf_data.iloc[:,[3,0,1,2]]
```

```python
Log_trf_data
```

```python
from sklearn.preprocessing import StandardScaler
scaleobj = StandardScaler()
scaled_data = scaleobj.fit_transform(Log_trf_data)
scaled_data = pd.DataFrame(scaled_data, columns=Log_trf_data.columns)
scaled_data.describe()
```

```python
from sklearn.cluster import KMeans 
cs_clus = KMeans()
l1 = []
for k in range(1,11):
    cs_clus = KMeans( n_clusters=k)
    ssd = np.abs(cs_clus.fit(Log_trf_data).score(Log_trf_data))
    l1.append(ssd)  
num = list(range(1,11))
plt.plot(num,l1,marker = "*")
plt.grid()
```

```python
cs_clus = KMeans( n_clusters=3)
ssd = np.abs(cs_clus.fit(Log_trf_data).score(Log_trf_data))
final_df = pd.concat([Log_trf_data,pd.Series(cs_clus.labels_)],axis=1)
final_df.head(5)
final_df[0].value_counts()
```
