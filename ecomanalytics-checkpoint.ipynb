```python
recency_date=dt.datetime(2017,12,19)
```

```python
r=ecom_final.groupby('CustomerID',as_index=False).agg({'date_of_purchase': lambda x:(recency_date -x.max()).days})
```

```python
m=ecom_final.groupby('CustomerID',as_index=False).agg({'Price' : lambda x :(x.max())})
```

```python
f=ecom_final.groupby('CustomerID',as_index=False).agg({'invoiceno': lambda x :(round(x.count()))})
```

```python
f
```

```python
m.drop('CustomerID',axis=1,inplace=True)
```

```python
f.drop('CustomerID',axis=1,inplace=True)
```

```python
rmf_scores=pd.concat([r,f,m],axis=1)
```

```python
rmf_scores.describe()
```

```python
rmf_scores.columns =['CustomerID','Recency','Frequency','Monetary']
```

```python
rmf_scores.head()
```

```python
rmf_scores['r_quartile'] = pd.qcut(rmf_scores['Recency'], 4, ['4','3','2','1'])
rmf_scores['f_quartile'] = pd.qcut(rmf_scores['Frequency'], 4, ['1','2','3','4'])
rmf_scores['m_quartile'] = pd.qcut(rmf_scores['Monetary'], 4, ['1','2','3','4'])
```

```python
rmf_scores['rmf'] =rmf_scores.r_quartile.astype(str)+ rmf_scores.f_quartile.astype(str) + rmf_scores.m_quartile.astype(str)
```

```python
rmf_scores=rmf_scores.astype({'r_quartile':'int64'})
```

```python
rmf_scores=rmf_scores.astype({'f_quartile':'int64'})
```

```python
rmf_scores=rmf_scores.dropna()
```

```python
rmf_scores=rmf_scores.astype({'m_quartile':'int64'})
```

```python
rmf_scores['rfmscore'] =rmf_scores[['r_quartile','f_quartile','m_quartile']].sum(axis=1)
```

```python
rmf_scores.shape
```

```python
rmf_scores.rfmscore.value_counts()
```

```python
loyalty=['bronze','silver','gold','platinum']
rmf_scores['loyalty_level']=pd.qcut(rmf_scores['rfmscore'],4,labels=loyalty)
```

```python
rmf_scores=rmf_scores.sort_values('rfmscore').reset_index()
```
